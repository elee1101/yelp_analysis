```python
import json
import pandas as pd
```

Sampling Review data

```python
review_path = "data/raw/yelp_academic_dataset_review.json"

review_data = []
user_ids = set()
business_ids = set()

sample_size = 50000

with open(review_path, 'r', encoding='utf-8') as f:
  for i, line in enumerate(f):
    if i >= sample_size:
      break

    record = json.loads(line)
    review_data.append(record)

    user_ids.add(record["user_id"])
    business_ids.add(record["business_id"])
  
review_df = pd.DataFrame(review_data)
review_df.to_csv("data/processed/reviews.csv", index=False)

print(f"reviews saved ({len(review_df)} rows)")
```

Match User and Review

```python
user_path = "data/raw/yelp_academic_dataset_user.json"

user_data = []

with open(user_path, 'r', encoding='utf-8') as f:
  for line in f:
    record = json.loads(line)

    if record["user_id"] in user_ids:
      user_data.append(record)

user_df = pd.DataFrame(user_data)
user_df.to_csv("data/processed/users.csv", index=False)

print(f"reviews saved ({len(user_df)} rows)")
```

Match Business and Review

```python
business_path = "data/raw/yelp_academic_dataset_business.json"

business_data = []

with open(business_path, 'r', encoding='utf-8') as f:
  for line in f:
    record = json.loads(line)

    if record["business_id"] in business_ids:
      business_data.append(record)

business_df = pd.DataFrame(business_data)
business_df.to_csv("data/processed/businesses.csv", index=False)

print(f"reviews saved ({len(business_df)} rows)")

```

Cleaning Data

- Removed problematic quotes from string fields using a DataFrame-wide transformation (ERROR: unterminated CSV quoted field)
- Dropped or excluded columns with complex structures (e.g., arrays, nested fields)

User Cleaning

```python
cols = [
    "user_id", "name", "review_count",
    "yelping_since", "useful", "funny",
    "cool", "fans", "average_stars"
]

user_clean = user_df[cols].copy()
```

```python
user_clean = user_clean.applymap(
    lambda x: str(x).replace('"', '') if isinstance(x, str) else x
)
```

```python
user_clean.to_csv(
    "data/processed/user_clean.csv",
    index=False,
    encoding="utf-8"
)
```

Business Cleaning

```python
cols = [
    "business_id", "name", "city", "state", "postal_code",
    "latitude", "longitude", "stars", "review_count", "is_open"
]

business_clean = business_df[cols].copy()
```

```python
business_clean = business_clean.applymap(
  lambda x: str(x).replace('"', '') if isinstance(x, str) else x
)
```

```python
business_clean.to_csv(
    "data/processed/business_clean.csv",
    index=False,
    encoding="utf-8"
)
```
